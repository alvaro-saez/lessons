## Wrangling methods

```python
import pandas as pd
import numpy as np
```

```python
def wrangling(df, year, col): #year should be an integer and col a string
    df = df[df[col].isin(year)]
    return df
```

```python
df_test = pd.read_csv("../../datasets/vehicles.csv")
year_test = [1985,1999]
col_test = "year"
```

```python
var = wrangling(df_test, year_test, col_test)
var
```
