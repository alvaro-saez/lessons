## Analysis methods

```python
import pandas as pd
import numpy as np
```

```python
def analisis(df, groupby_col, agg_col):
    df = df.groupby([groupby_col])[agg_col].sum()
    df.to_csv("../data/vehicles.csv")
    message = "/--------PIPELINE SUCCESSFULY EXECUTED-------/"
    return message
```

```python
def wrangling(df, year, col): #year should be an integer and col a string
    df = df[df[col].isin(year)]
    return df


df_test = pd.read_csv("../../datasets/vehicles.csv",low_memory=False)
year_test = [1985,1999]
col_test = "year"

var = wrangling(df_test, year_test, col_test)
var
```

```python
groupby_col_test = "year"
agg_col_test = "co2"
analisis(var, groupby_col_test, agg_col_test)
```
