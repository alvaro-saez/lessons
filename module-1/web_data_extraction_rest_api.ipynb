```python
# Import libraries
import pandas as pd
import requests   #!conda install requests

# Pandas display options
pd.set_option('display.max_columns', None)
```

---

```python
response = requests.get('https://jsonplaceholder.typicode.com/todos') # realizamos la request para solicitar la info
print(type(response)) 
```

### HTTP Response

[Boring Reference](https://en.wikipedia.org/wiki/List_of_HTTP_status_codes)

[Funny Reference](https://http.cat/)

```python
status = response.status_code
status
```

```python
content = response.content
content
```

```python
json_data = response.json() #lo parseamos en un json

print(type(json_data))
print(len(json_data))
print(type(json_data[0]))
#json_data[0].keys()

json_data
```

```python
# Other attributes/methods: .headers, .links, .cookies

headers = response.headers
headers
```

```python
# But we like DataFrames

df = pd.DataFrame(json_data)
df
```

---

```python
# Main End-Point
end_point = 'https://api.github.com/'
# Body
par1 = 'repos/'
par2 = 'ih-datapt-mad/'
par3 = 'dataptmad1121_labs/'
par4 = 'pulls'   # https://docs.github.com/en/rest/reference/pulls
par5 = '?state=open'
par6 = "&perpage=100"
par7 = "&page=1"
```

```python
pulls_response = requests.get(end_point + par1 + par2 + par3 + par4 + par5 + par6 + par7)
pulls_json = pulls_response.json()
print(type(pulls_json))
print(len(pulls_json))
print(pulls_json[0].keys())
```

```python
df_pulls = pd.DataFrame(pulls_json)
df_pulls.head()
```

---

```python
# The best way is with a method called: .json_normalize --> APLANA EL JSON

df_flat = pd.json_normalize(pulls_json)#.T.reset_index(drop=True)
print(df_flat.info())
df_flat
```
